```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# demo code

x = [1,2,3]
y= [4,5,6]

plt.plot(x,y) # {(x1, y1), (x2, y2), (x3, y3)}
plt.grid()
plt.show()
```

```python
# Univariate - Numerical

data = {
    'Salary': [1000, 500, 20000, 234, 5067, 1409, 2344, 6789,3421, 100000, 56078, 10]
}

df = pd.DataFrame(data)
df.shape
```

```python
# line graph

plt.plot(df['Salary'])
plt.show()
```

```python
# customized line graph
plt.plot(df['Salary'], linewidth='2', color='blue', marker='o', linestyle=':')
plt.show()
```

```python
# histogram

plt.hist(df['Salary'], bins=5,color='red')
plt.show()
```

```python
# boxplot
plt.boxplot(df['Salary'])
plt.show()
```

```python
# Univariate - Categorical

df['dept'] = ['HR', 'Finance', 'IT', 'Management']*3
df
```

```python
df.info()
```

```python
# pie chart

count = df['dept'].value_counts()
count

plt.pie(count, labels=count.index, autopct="%1.1f", explode=[0.3, 0, 0.3, 0])
plt.show()
```

```python
# countplot

plt.bar(count.index, count)
plt.show()
```

```python
# bivariate - numerical and numerical
df['Age'] = [45,23,54, 21, 22, 19]*2
df
```

```python
# scatterplot
plt.scatter(df['Age'], df['Salary'], color='blue')
plt.show()
```

```python
# line graph
sorted_age = df.sort_values('Age')
plt.plot(sorted_age['Age'], df['Salary'])
plt.show()
```

```python
# bar graph
plt.bar(sorted_age['Age'], df['Salary'], color='red')
plt.show()
```
