```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# demo code

x = [1,2,3]
y= [4,5,6]

plt.plot(x,y) # {(x1, y1), (x2, y2), (x3, y3)}
plt.grid()
plt.show()
```

```python
# Univariate - Numerical

data = {
    'Salary': [1000, 500, 20000, 234, 5067, 1409, 2344, 6789,3421, 100000, 56078]
}

df = pd.DataFrame(data)
df.shape
```

```python
# line graph

plt.plot(df['Salary'])
plt.show()
```

```python
# customized line graph
plt.plot(df['Salary'], linewidth='2', color='blue', marker='o', linestyle=':')
plt.show()
```
